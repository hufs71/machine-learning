### Wine using DT

```python
import pandas as pd

white_url = "https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-white.csv"
red_url = "https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-red.csv"

red_wine = pd.read_csv(red_url, sep=';')
white_wine = pd.read_csv(white_url, sep=';')
```

```python
red_wine.head()
```

```python
white_wine.head()
```

```python
white_wine.columns
```

```python
red_wine['color'] = 1
white_wine['color'] = 0

wine = pd.concat([red_wine, white_wine])
wine.info()
```

```python
wine.head()
```

```python
wine['quality'].unique()  # 등급
```

```python
import plotly.express as px

fig = px.histogram(wine, x='quality')
fig.show()
```

```python
wine['quality'].value_counts()
```

* red/white 와인별로 등급 histogram

```python
fig = px.histogram(wine, x='quality', color='color')
fig.show()  # 1: red, 0: white
```

#### red/white wine 분류기

```python
X = wine.drop(['color'], axis=1)  # feature
y = wine['color']  # label
```

* train/test split

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
np.unique(y_train, return_counts=True)
```

--> 대충 3900개: 1300개로 분리됨

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Histogram(x=X_train['quality'], name='Train'))
fig.add_trace(go.Histogram(x=X_test['quality'], name='Test'))

fig.update_layout(barmode='overlay')  # 겹쳐서 그리기
fig.update_traces(opacity=0.75)
fig.show()
```

* DT 훈련

```python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
# accuracy 확인
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc: ', accuracy_score(y_train, y_pred_tr))
print('Test Acc: ', accuracy_score(y_test, y_pred_test))
```

```python
accuracy_score(y_train, y_pred_tr)  # (참값, 예측값)
```

```python
accuracy_score(y_test, y_pred_test)
```

### 와인 맛에 대한 분류 - 이진 분류

* quality column 이진화

```python
# 5등급보다 크면 1
wine['taste'] = [1. if grade>5 else 0. for grade in wine['quality']]
wine.head()
```

```python
X = wine.drop(['taste'], axis=1)
y = wine['taste']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

wine_tree = DecisionTreeClassifier(max_depth=4, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc: ', accuracy_score(y_train, y_pred_tr))
print('Test Acc: ', accuracy_score(y_test, y_pred_test))
```
